```python
# 11. Visualize the student mark using live chart 
# 15.Display "X"-axis and "Y"-axis label


x=ds['Marks']
y=ds['Roll Number']
plt.plot(y,x)
plt.title("Student Marks")
plt.ylabel("Marks")
plt.xlabel("Roll Number")
plt.show()
```

```python
# 12. Display the name and mark information using bar chat 

x=ds['Marks']
y=ds['Roll Number']
plt.bar(y,x)
plt.title("Student Marks")
plt.ylabel("Marks")
plt.xlabel("Roll Number")
plt.show()
```
